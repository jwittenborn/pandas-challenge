```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
NumPlayers = len(purchase_data["SN"].value_counts())
print(f"Player Count: {NumPlayers}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Run basic calculations to obtain number of unique items, average price, etc.

NumItems = len(purchase_data["Item ID"].value_counts())

AvgPrice = purchase_data["Price"].mean()

NumPurchases = purchase_data["Purchase ID"].count()

TotRev = purchase_data["Price"].sum()

#Create a summary data frame to hold the results
PurchaseStats_df = pd.DataFrame({
    "Number of Unique Items":[NumItems],
    "Average Price":[AvgPrice],
    "Number of Purchases":[NumPurchases],
    "Total Revenue":[TotRev]
})

#Optional: give the displayed data cleaner formatting
#Display the summary data frame
PurchaseStats_df.style.format({
    'Average Price':"${:,.2f}",
    'Total Revenue': '${:,.2f}'
})

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#group by gender

GenderGrp = purchase_data.groupby("Gender")
GenderGrp.count()


UniquePlayers = GenderGrp.nunique()["SN"]
#print(UniquePlayers)

PctPlayers = UniquePlayers / NumPlayers * 100
#print(PctPlayers)

GenderDemog_df = pd.DataFrame({
    "Total Count":UniquePlayers,
    "Percentage of Players":PctPlayers
})
#GenderDemog_df


GenderDemog_df.style.format({
    'Total Count':"{:,.0f}",
    'Percentage of Players': '{:,.2f}%'
})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
MinAge = purchase_data["Age"].min()
MinAge

MaxAge = purchase_data["Age"].max()
print(f"Minimum age: {MinAge}, Maximum age: {MaxAge}")

ageBins = [0,9.9999,14.9999,19.9999,24.9999,29.9999,34.9999,39.9999,199]
ageBinNames = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#cut data
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], ageBins, labels=ageBinNames, include_lowest=True)
purchase_data

#create new DF grouped by Age Group
purchase_data_grp = purchase_data.groupby("Age Group")
purchase_data_grp.max()

#number of players per age group
NumPlayByAge = purchase_data_grp["SN"].nunique()
NumPlayByAge

#percentage of players in each age group
PctPlayByAge = NumPlayByAge / NumPlayers * 100
PctPlayByAge

AgeTable_df = pd.DataFrame({
    "Total Count":NumPlayByAge,   
    "Percentage of Players":PctPlayByAge
})


AgeTable_df.style.format({
    'Percentage of Players':"{:,.2f}%",
})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
BySpender = purchase_data.groupby("SN")

TotPurchaseCountPP = BySpender["Purchase ID"].count()
TotPurchaseCountPP

#avg purchase price per person
AvgPurchPricePP = BySpender["Price"].mean()
AvgPurchPricePP

#total purchase value per person
TotPurchValuePP = TotPurchaseCountPP*AvgPurchPricePP
TotPurchValuePP

spenders = pd.DataFrame({
    "Purchase Count": TotPurchaseCountPP,
    "Average Purchase Price": AvgPurchPricePP,
    "Total Purchase Value":TotPurchValuePP
})
spenders

SpendersSort = spenders.sort_values(["Total Purchase Value"], ascending=False).head(5)
SpendersSort

SpendersSort.style.format({
    'Purchase Countt':"{:,.0f}%",
    'Average Purchase Price':"${:,.2f}",
    'Total Purchase Value':"${:,.2f}",
})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Items = purchase_data[["Item ID","Item Name","Price"]]

ItemsByID_Name = Items.groupby(["Item ID", "Item Name"])

PurchaseCount = ItemsByID_Name["Price"].count()
PurchaseCount

TotalPurchaseValue = ItemsByID_Name["Price"].sum()
TotalPurchaseValue

ItemPrice = ItemsByID_Name["Price"].mean()
ItemPrice

popularItems = pd.DataFrame({
    "Purchase Count": PurchaseCount, 
    "Item Price": ItemPrice,
    "Total Purchase Value":TotalPurchaseValue})
popularItems

popularItemsSort = popularItems.sort_values(["Purchase Count"], ascending=False).head(5)
popularItemsSort
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profitableItemsSort = popularItems.sort_values(["Total Purchase Value"], ascending=False).head(5)
profitableItemsSort

profitableItemsSort.style.format({
    'Purchase Countt':"{:,.0f}%",
    'Item Price':"${:,.2f}",
    'Total Purchase Value':"${:,.2f}",
})
```
